```python
# Gets historical Stock price data from yahoo finance. 
# Imputs: ticker - the stock ticker. String. 1-4 characters.
#         time_period - the time, in seconds of history back from the present that data should be retrived. Int.
#         interval - how often the data should be aggregated. Str. either "1d", "1w", or "1m"
# output: a pandas dataframe.
def get_historical_data(ticker, time_period, interval):
    period2 = int(time.mktime(datetime.date.today().timetuple()))
    period1 = int(time.mktime(datetime.date.today().timetuple())) - time_period
    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
    return pd.read_csv(query_string) 
```

```python
ticker = 'TSLA'
period = 3600*24*7*26
interval = '1d'
df = get_historical_data(ticker, period, interval)
```

```python
def get_options_data(ticker, time_period, interval):
    t = int(time.mktime(datetime.date.today().timetuple()))
    period2 = str(datetime.datetime.fromtimestamp(t).strftime("%Y-%m-%d"))
    period1 = str(datetime.datetime.fromtimestamp(t- time_period).strftime("%Y-%m-%d"))
    query_string = f'https://www.cboe.com/us/options/market_statistics/historical_data/download/class/?reportType=volume&volumeType=sum&volumeAggType=daily&symbolType=underlying&symbol={ticker}&startDate={period1}&endDate={period2}&exchanges=EDGX&exchanges=C2&exchanges=BATS&exchanges=CBOE'
    return pd.read_csv(query_string) 
```
